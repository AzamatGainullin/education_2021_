```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv', sep=';', error_bad_lines=False, encoding='windows-1251', nrows=1000000) # nrows=1000000
```

```python
df.info()
```

```python
def add_patient_id(df):
    df['PATIENT_ID'] = df['DOCUMENT_ID']
    df5=df[df.COL=='ид_пациента']
    df6 = df5.VAL.value_counts()
    df7=df6[df6>1]
    for i in df7.index:
        df_temp=df[df.VAL==i]
        spisok=df_temp.PATIENT_ID.values
        for item in spisok[1:]:
            df['PATIENT_ID'] = df['PATIENT_ID'].replace([int(item)],int(i))
    return df
```

```python
df = add_patient_id(df)
```

```python
zz = df[(df.COL=='вид_вакцинации')]
```

```python
pd.pivot_table(zz,index=["VAL"], values=["LINE_ID", 'PATIENT_ID'], aggfunc={"LINE_ID":len, 'PATIENT_ID':lambda x:len(np.unique(x))}, fill_value=0)
```
